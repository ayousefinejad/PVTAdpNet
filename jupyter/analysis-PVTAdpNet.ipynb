```python
import matplotlib.pyplot as plt

def plot_losses(train_loss, val_loss):
    epochs = range(1, len(train_loss) + 1)
    plt.figure(figsize=(12, 6))
    plt.plot(epochs, train_loss, 'b-o', label='Train Loss')
    plt.plot(epochs, val_loss, 'r-o', label='Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('PVTAdapNet Train and Validation Loss')
    plt.legend()
    plt.grid(True)
    plt.show()

def plot_jaccard(train_jaccard, val_jaccard):
    epochs = range(1, len(train_jaccard) + 1)
    plt.figure(figsize=(12, 6))
    plt.plot(epochs, train_jaccard, 'b-o', label='Train Jaccard Index')
    plt.plot(epochs, val_jaccard, 'r-o', label='Validation Jaccard Index')
    plt.xlabel('Epochs')
    plt.ylabel('Jaccard Index')
    plt.title('PVTAdapNet Train and Validation Jaccard Index')
    plt.legend()
    plt.grid(True)
    plt.show()

def plot_f1(train_f1, val_f1):
    epochs = range(1, len(train_f1) + 1)
    plt.figure(figsize=(12, 6))
    plt.plot(epochs, train_f1, 'b-o', label='Train F1 Score')
    plt.plot(epochs, val_f1, 'r-o', label='Validation F1 Score')
    plt.xlabel('Epochs')
    plt.ylabel('F1 Score')
    plt.title('PVTAdapNet Train and Validation F1 Score')
    plt.legend()
    plt.grid(True)
    plt.show()

def plot_recall(train_recall, val_recall):
    epochs = range(1, len(train_recall) + 1)
    plt.figure(figsize=(12, 6))
    plt.plot(epochs, train_recall, 'b-o', label='Train Recall')
    plt.plot(epochs, val_recall, 'r-o', label='Validation Recall')
    plt.xlabel('Epochs')
    plt.ylabel('Recall')
    plt.title('PVTAdapNet Train and Validation Recall')
    plt.legend()
    plt.grid(True)
    plt.show()

def plot_precision(train_precision, val_precision):
    epochs = range(1, len(train_precision) + 1)
    plt.figure(figsize=(12, 6))
    plt.plot(epochs, train_precision, 'b-o', label='Train Precision')
    plt.plot(epochs, val_precision, 'r-o', label='Validation Precision')
    plt.xlabel('Epochs')
    plt.ylabel('Precision')
    plt.title('PVTAdapNet Train and Validation Precision')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage:
# plot_losses(train_loss, val_loss)
# plot_jaccard(train_jaccard, val_jaccard)
# plot_f1(train_f1, val_f1)
# plot_recall(train_recall, val_recall)
# plot_precision(train_precision, val_precision)
```
